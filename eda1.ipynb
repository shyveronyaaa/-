```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df.info()
```

```python
fig,ax = plt.subplots(ncols=2,nrows=2,figsize=(10,10))
sns.histplot(df,x='price',hue='transmition',ax=ax[0][0])
sns.histplot(df,x='price',hue='fuel',ax=ax[0][1])
sns.histplot(df,x='price',hue='driver',ax=ax[1][0])
sns.histplot(df,x='price',hue='brand',ax=ax[1][1])
fig.tight_layout(pad=5)
```
